#**Data Toolkit**

1. What is NumPy, and why is it widely used in Python?

   -  NumPy is a Python library for numerical computing. It provides support for large,
    multi-dimensional arrays and matrices, along with mathematical functions to operate on them
    efficiently. It’s widely used because of its speed, ease of integration with other libraries like Pandas
    and Scikit-learn, and its vectorized operations that replace explicit loops.

2. How does broadcasting work in NumPy?

    - Broadcasting allows NumPy to perform arithmetic operations on arrays of different shapes by
    automatically expanding the smaller array to match the larger array’s shape.


3. What is a Pandas DataFrame?
  
   - A Pandas DataFrame is a two-dimensional, size-mutable, and heterogeneous data structure in
    Python that stores data in labeled rows and columns, similar to a spreadsheet or SQL table.

4. Explain the use of the groupby() method in Pandas.

    - The groupby() method in Pandas is used to group data based on one or more keys, allowing
    aggregation, transformation, and filtration of grouped data efficiently

5. Why is Seaborn preferred for statistical visualizations?
   - Seaborn is preferred because it provides a high-level interface for creating attractive and
    informative statistical plots with fewer lines of code, and integrates well with Pandas data
    structures.

6. What are the differences between NumPy arrays and Python lists?
   - NumPy arrays are more memory-efficient, support vectorized operations, and have fixed types,
    while Python lists are more flexible but slower for numerical computations.


7. What is a heatmap, and when should it be used?

   - A heatmap is a graphical representation of data where individual values are represented by
    color. It is used to visualize correlations, frequency, or intensity patterns in a dataset

8. What does the term “vectorized operation” mean in NumPy?

   - Vectorized operations mean performing operations on entire arrays without writing explicit loops,
    making computations faster and more concise.

9. How does Matplotlib differ from Plotly?

   - Matplotlib is static and suited for simple visualizations, while Plotly creates interactive and
    web-based visualizations.


10. What is the significance of hierarchical indexing in Pandas?

    - Hierarchical indexing allows multiple levels of index on a DataFrame, enabling more complex
    data representation and easier subsetting

11. What is the role of Seaborn’s pairplot() function?

     - The pairplot() function visualizes pairwise relationships between numerical variables in a
      dataset, showing both scatter plots and distributions.

12. What is the purpose of the describe() function in Pandas?
    
     - The describe() function provides summary statistics like mean, median, count, and standard
      deviation for numerical columns.

13. Why is handling missing data important in Pandas?

     -  Handling missing data ensures accurate analysis, as missing or NaN values can distort
       statistical results and cause errors.


14. What are the benefits of using Plotly for data visualization?

     -  Plotly offers interactive, web-based, and highly customizable visualizations that are easy to
     share and explore

15. How does NumPy handle multidimensional arrays?

     - NumPy handles multidimensional arrays using the ndarray object, allowing efficient operations
      across dimensions using indexing and slicing.


16. What is the role of Bokeh in data visualization?
    
    -  Bokeh is used to create interactive, web-ready visualizations that can handle large and
      streaming datasets efficiently.


17. Explain the difference between apply() and map() in Pandas.
  
     - The apply() function can be used on DataFrames to apply a function along an axis, while map()
      is used to transform each element in a Series.


18. What are some advanced features of NumPy?

      -  Advanced features include broadcasting, masked arrays, linear algebra operations, FFT,
       random number generation, and memory mapping of large files.

19. How does Pandas simplify time series analysis?

      - Pandas provides built-in support for datetime indexing, resampling, shifting, and rolling
       operations, making time series manipulation easy.

20. What is the role of a pivot table in Pandas?

       - Pivot tables summarize data by reorganizing and aggregating it based on specified columns,
         useful for quick analysis.

21. Why is NumPy’s array slicing faster than Python’s list slicing?

      -  NumPy’s arrays store elements in contiguous memory blocks and use vectorized C-based
        operations, making slicing faster than Python lists.

22. What are some common use cases for Seaborn?

      - Seaborn is commonly used for correlation heatmaps, distribution plots, categorical comparisons,
              and regression analysis.

1. How do you create a 2D NumPy array and calculate the sum of each row?


```python
import numpy as np

# Create a 2D NumPy array
arr = np.array([[1, 2, 3], [4, 5, 6]])

# Calculate the sum of each row
row_sum = arr.sum(axis=1)

# Display the result
print(row_sum)
```

2. Write a Pandas script to find the mean of a specific column in a DataFrameA.

```python
import pandas as pd

# Create a sample DataFrame
data = {'A': [10, 20, 30, 40, 50]}
df = pd.DataFrame(data)

# Find the mean of a specific column 'A'
mean_value = df['A'].mean()

# Display the result
print("Mean of column A:", mean_value)
```

3. Create a scatter plot using MatplotlibA.

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [10, 15, 20, 25, 30]

# Create a scatter plot
plt.scatter(x, y, color='blue', marker='o')

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot Example')

# Display the plot
plt.show()
```

4. How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap.

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Create a sample DataFrame
data = pd.DataFrame({
    'A': [1, 2, 3, 4, 5],
    'B': [5, 4, 3, 2, 1],
    'C': [2, 3, 4, 5, 6]
})

# Calculate the correlation matrix
corr_matrix = data.corr()

# Visualize the correlation matrix using a heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix Heatmap')
plt.show()
```

5. Generate a bar plot using PlotlyA.

```python
import plotly.express as px
import pandas as pd

# Create a sample DataFrame
data = pd.DataFrame({
    'Fruits': ['Apple', 'Banana', 'Mango', 'Orange'],
    'Count': [10, 15, 7, 12]
})

# Generate a bar plot
fig = px.bar(data, x='Fruits', y='Count', title='Fruit Count', color='Fruits')

# Display the plot
fig.show()
```

8.  Create a line plot with multiple lines using Matplotlib.

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y1 = [10, 20, 25, 30, 35]
y2 = [5, 15, 20, 22, 28]

# Create a line plot with multiple lines
plt.plot(x, y1, label='Line 1', color='blue', marker='o')
plt.plot(x, y2, label='Line 2', color='red', marker='s')

# Add labels, title, and legend
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Multiple Line Plot')
plt.legend()

# Display the plot
plt.show()
```

10. A Create a histogram using Seaborn to visualize a distributionA.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
data = [1, 2, 2, 3, 3, 3, 4, 4, 5, 5, 5, 5]

# Create a histogram to visualize the distribution
sns.histplot(data, bins=5, kde=True, color='skyblue')

# Add title and labels
plt.title('Distribution Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Display the plot
plt.show()
```
